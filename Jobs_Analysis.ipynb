Data Cleaning, Processing and Exploration

```python
mydata.info()
```

```python
mydata.replace('-1', np.nan, inplace=True)
mydata.replace(-1, np.nan, inplace=True)
mydata.replace(-1.0, np.nan, inplace=True)
```

```python
mydata.isnull().sum()
```

```python
mydata['min_size'] = mydata['Size'].str.split(' ').str[0]
mydata
```

```python
mydata['Founded']=mydata['Founded'].fillna(0.0).astype('float')
mydata['Founded'] = mydata['Founded'].astype(int)
mydata
```

```python
mydata['Company Name'] = mydata['Company Name'].str.split('\n').str[0]
mydata
```

```python
mydata["State"] = mydata["Location"].agg(lambda x: x[-2:])
mydata["City"] = mydata["Location"].agg(lambda x: x[:-4])
```

```python
mydata['Headquarters'] = mydata['Headquarters'].str.split(',').str[0]
mydata
```

```python
mydata = mydata.rename(columns={'Salary Estimate': 'Salary'})
```

```python
mydata.Salary.isnull().sum()
```

```python
mydata[mydata['Salary'].isnull()]
```

```python
df_salary = mydata['Salary'].str.split("-",expand=True,)

minimum_salary = df_salary[0]
minimum_salary = minimum_salary.str.replace('K',' ')


maximum_salary = df_salary[1].str.replace('(Glassdoor est.)', ' ')

maximum_salary = maximum_salary.str.replace('(', ' ')
maximum_salary = maximum_salary.str.replace(')', ' ')
maximum_salary = maximum_salary.str.replace('K', ' ')
maximum_salary = maximum_salary.str.replace('Employer est.', ' ')
maximum_salary = maximum_salary.str.replace('Per Hour', ' ')

maximum_salary = maximum_salary.str.replace('$', ' ').fillna(0).astype(int)
minimum_salary = minimum_salary.str.replace('$', ' ').fillna(0).astype(int)

mydata['Salary_minimum'] = minimum_salary
mydata['Salary_maximum'] = maximum_salary

mydata
```

```python
mydata['Salary_average'] = (mydata['Salary_maximum']+mydata['Salary_minimum'])/2
mydata['Salary_average']=mydata['Salary_maximum'].fillna(0.0).astype('float')
mydata['Salary_average'] = mydata['Salary_average'].astype(int)
mydata
```

```python
mydata.drop(['Salary','Job Description','Location','Size','Revenue','Easy Apply', 'Competitors'], axis=1, inplace= True)
mydata
```
